````python
# write a program to print 1 to 20 numbers using for loop
for i in range(1,21):
    print (i)
````

````python
# write a program to iterate a string using for loop
info =("welcome to python program")
for char in info:
    print(char)
````

````python
# write a program to print given numbers in between  50 to 100 using for loop
for n in range (50,101):
    if (n%2 == 0):
       print(n,end=",")
````

````python
#write a program to print odd numbers in between 50 to 100 using for loop
for n in range(50,101):
    if(n%2 == 1):
        print (n,end=",")
````

````python
# write a program to find vowels from string using for loop
data = "python programming"
for char in data:
    if (char in "aeiouAEIOU"):
        print(char)
````

````python
# write a program to find constonants using for loop
data = "python programming"
for char in data:
    if(char not in "aeiouAEIOU"):
        print(char)
````

````python
# write a program to find the sum of all numbers in between 1 to 100 using for loop
total = 0

for i in range(1, 101):
    total += i

print("The sum of all numbers from 1 to 100 is:", total)
````

````python
#write a program to find the sum of all even numbers 1 to 50 using for loop

total = 0

for i in range(2, 51, 2):
    total += i

print("The sum of all even numbers from 1 to 50 is:", total)
````

````python
# write a program to find the sum of all odd numbers in between 1 to 50 numbers using for loop

total = 0

for i in range(1, 51, 2):
    total += i

print("The sum of all odd numbers from 1 to 50 is:", total)
````
